# Set Operations 

```python

set1 = {1,2,4,3}
set2 = {3,4,5,6}
```

```python
# Union set  operation
set1.union(set2)
```

```python
# Union set  operation
print(set1 | set2)
```

```python
# Intersection set operation
print(set1 & set2)
set1.intersection(set2)
```

```python
# Difference set operation
print(set1 - set2)
set1.difference(set2)
```

```python
# Symmmetric Difference set operartion
print(set1 ^ set2)
set1.symmetric_difference(set2)
```

```python
 sub = {1,3}
```

```python
sub.issubset(set1)
```

```python
set1.issuperset(sub)
```
