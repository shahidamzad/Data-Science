# List comprehension
###### is a concise and effient way to create list in python . It allow you to genrete list in a single line of code , making code more readable and pythonic

```python
table=[]
```

```python
for i in range (1,11):
    print(f" 2 x {i} = {2*i}")
    table.append(2*i)
```

```python
t = [2*i for i in range(1,11)]
```

```python
t
```

```python
# for filtering even number 

numbers = [2*i for i  in range(1,11) if i % 2 == 0 ]

```

```python
numbers
```

```python
# if-else condis=tion for replacing even with even or add numbers

num = ["Even" if i%2 == 0 else "odd " for i in range( 1,11)]
```

```python
num
```

```python
# nested loop = creating pair of list , useally used in matrix

pairs = [(x, y) for x in range(2) for y in range(3)]
```

```python
pairs
```

```python
# nested loop for i matrix

matrix = [[1 if i == j else 0 for j in range(3)] for i in range(3)]
```

```python
matrix
```
