# Working with binary file 

```python
# binary file reading
with open("image.png", "rb" ) as file :
    data = file.read()
    print(data)
```

```python
# write to a binary file

with open("new_image.jpg" , "wb") as file:
    write = file.write(data)
    print(write)
```

```python
# Appending to a binary file

file = open("example.txt" , "a")
file.write( "\n thiss is an additional line")
file.close()
```

```python
# ussing with  statement  (best practice)
with open("example.txt" , "r") as file:
    content = file.read()
    print(content)
```

```python
# cheaking file exists or not

import os
if os.path.exists("example.txt"):
    print("File exits!")
else:
    print("File dose not exits")
```

```python
# deleting a file 

import os
if os.path.exists("example.txt"):
    os.remove("example.txt")
    print("File deleted")
else:
    print("File dose not exists")
```
