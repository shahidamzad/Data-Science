# python in JSON Module - working with JSON data
##### json is (javascript object notation) lightweight data formate used for data exchanged between server and application .  (API , web applications  and configration) 

```python
import json
```

```python
obj = {"name":"shahid", "company":"tcs", "salary":65000}
s = json.dumps(obj)   # convets python object to JSON string
type(s)
```

```python
d = json.loads('{"name":"shahid", "company":"tcs", "salary":65000}')  # covert JSON  strings to python object
```
