# Loop in python

```python
#for loop =  like a list , tuple and string

fruits = ["apple", 'banana', 'orange']
for fruit in fruits:
    print (fruit)

```

```python
l = [1,4,4,57,8765,85]
```

```python
# use range()

for i in range(5):
    print(i)
```

```python
for i in range (0,11):
    print(i)
```

 # While loop
 #### run as long as condition is true

```python
count = 0 
while count < 3:
    print(count)
    count +=1
```

```python
count = 2
while count < 21:
    print(count)
    count += 2
```

```python
i = 0
while i < 12:
    print(i)
    i = i + 1
```

# LOOP control statement 

#### break --exits the loop 
#### continue -- skips the next loop iteration
#### pass -- does nothing (used as a placeholder)

```python
for item in l:
    pass  # dont do anything
print("hello")
```

```python
l = ['banana','apple','ornage']
for item in l:
    print(item)
```

```python
l = ['banana','apple','ornage']
for item in l:
    if(item == 'apple'): 
        break  # stop loop right here 
print(item)
```

```python
l = ['banana','apple','ornage']
for item in l:
    if(item == 'apple'): 
        continue  # skip this particular iteration of the loop or print next item
print(item)
```
