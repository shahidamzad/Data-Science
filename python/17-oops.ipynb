```python
class Employe:
    company = "HP"  # class attribute 

    def __init__(self, name, salary):   # constructor  = take argument
        self.name = name
        self.salary = salary
    

    def printDetails(self):
        print(f"The company of {self.name} is {self.company} with salary ${self.salary}")
      
    @staticmethod  # Dose not use class or instance attribte
    def printTime():
        print(f"The time is now ")

    @classmethod  # work with class attribute 
    def printClassDetails(cls):
        print("The comapany is {cls.company}")
         

e = Employe("shahid", 50000 , )  # object 
e.printDetails()
e.printTime()
e.printClassDetails()
```

```python
# Encapsulation  = Encapsulation means wrapping data and methods together and restricting direct access.
class BankAccount:
    def __init__(self):
        self.__balance = 0   # private variable

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

```

```python
acc = BankAccount()
acc.deposit(1000)
print(acc.get_balance())
```

```python
# Inharitence  = Inheritance allows one class to acquire properties of another class.
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")
```

```python
d = Dog()
d.sound()
```

```python
# Plymorphism = Polymorphism means same method name, different behavior.
class Bird:
    def fly(self):
        print("Bird can fly")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")
```

```python
b = Bird()
p = Penguin()

b.fly()
p.fly()
```

```python
# Abstraction = Abstraction hides implementation details and shows only essential features.
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
class Car(Vehicle):
    def start(self):
        print("Car starts with key")

```

```python
c = Car()
c.start()
```

```python
# OOP Basics in Python

# Define a class
class Student:
    # Constructor
    def __init__(self, name, age, course):
        self.name = name        # attribute
        self.age = age          # attribute
        self.course = course   # attribute

    # Method
    def introduce(self):
        return f"My name is {self.name}, I am {self.age} years old and learning {self.course}."

# Create objects
student1 = Student("Alice", 22, "Data Science")
student2 = Student("Bob", 25, "Python")

# Access attributes
print(student1.name)
print(student2.course)

# Call methods
print(student1.introduce())
print(student2.introduce())
```
