```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
s.sum()
```

```python
b = np.sum(s,axis = 1)
b==45
```

```python
b = np.sum(s,axis = 1)
for i in b:
    if i !=45:
        print("sodoku is not valid")
        break
    else:
        print("For row is valid ")
```

```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
# Get the shape of the matrix
data.shape
```

```python
# find the average  age sum of the matrix
np.mean(data[:,0])
```

```python
# Extract math markes of all student 
data[:,1]
```

```python
# find the highest science marks
np.max(data[:,2])
```

```python
# Get details of the student who scored more then 90 in maths 
data[data[:,2] > 90]
```

```python
#increase maths marks all of student by 4
data[:,1 ] + 5
```

```python
# find how many student are younger then 19
len(data[ data[:,0] < 19])
```

```python
# calculeter  the average marks in each subject ( colums-wise mean)
np.mean(data[:,1:],axis = 0)
```

```python
# Get the data of student scored at least 80 in bath subjects
data[(data[:,1]>=80) & (data[:,2] >=80)]
```

```python
# Replace all science marks  < 75 with 0

data[:,2][data[:,2] < 75] = 0
data
```
