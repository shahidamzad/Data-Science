# NUMPY INDEXING AND SLICING OF MATRIX

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[5] # for row
```

```python
arr[0,0] # for single element
```

```python
arr[0,4]
```

```python
slice = arr[0:2,1:3]
```

```python
slice
```

```python
arr[3:,3:]
```

```python
# for columns
arr[:,3]
```

# BOOLEAN INDEXING

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr %2 == 0
```

```python
bool_index
```

```python
arr = arr[bool_index]
arr
```
