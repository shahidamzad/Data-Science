# Adding to list by using python



```python
# using python list 

import numpy as np
import time

size = 10_00_000
l1 = list(range(size))
l2 = list(range(size))

start = time.time()

add = [x+y for x,y in zip (l1,l2)]
end = time.time()
print(add[0:10])
print(end - start)
```

# Using  numpy  array
#####  NumPy stores data in a contiguous memory block, making access faster than lists.ðŸ”¹ shape shows the dimensions of an array.


```python
import numpy as np
import time

size = 10_00_000

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()

add = l1+l2
end = time.time()
print(add[0:10])

print(end - start)
```

# numpy vectorization - No more loop
#### NumPy avoids loops by applying operations to entire arrays at once using SIMD
#### (Single Instruction, Multiple Data) and other low-level optimizations. SIMD is a
#### CPU-level optimization provided by modern processors.

```python
# python list loop
list1=[2,3,5,6,7]

list_square = [x**2 for x in list1]
print(list_square)


# numpy (vectorization)
arr1=np.array([2,5,6,7,3])
numpy_square = arr1**2
print(numpy_square)
```
